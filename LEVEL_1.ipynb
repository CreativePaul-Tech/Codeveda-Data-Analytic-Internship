# TASK 1: DATA CLEANING AND PREPROCESSING

```python
import pandas as pd
# Load the dataset
df = pd.read_csv('1) iris.csv')
# look at the first 5 rows 
df.head()
```

```python
import os
print(os.listdir())
```

```python
#check for missing values in each column
print ("missing values:")
print (df.isnull().sum())

#check for exact duplicate rows
print("\nDuplicate Rows:")
print (df.duplicated().sum())
```

```python
#Remove the duplicate rows and save it as a new, clean dataset
df_cleaned = df.drop_duplicates()

#verify the duplicates are completely gone
print("Duplicates remaining:", df_cleaned.duplicated().sum())

#check the final size of our clean dataset
print ("Total rows after cleaning:", len(df_cleaned))
```

# Task 2: EXPLORATORY DATA ANALYSIS (EDA)

```python
#import the visualization tools
import matplotlib.pyplot as plt
import seaborn as sns

#create a histogram to see the distribution of sepal lengths
sns.histplot(data=df_cleaned, x='sepal_length', bins=15)

#Add a title so we know what we are looking at
plt.title('Distribution of sepal length')

#Tell python to display the chart
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#make the chart a bit larger
plt.figure(figsize=(10, 6))

#create a boxplot for all numerical columns in our dataset
sns.boxplot(data=df_cleaned)
#Add a title
plt.title('Boxplot of Iris Features')

#Display the chart
plt.show()
```

```python
#make the chart a bit larger
plt.figure(figsize=(8,6))

#create a scatter plot comparing our two chosen columns
sns.scatterplot(data=df_cleaned, x='sepal_length', y='petal_width')

#Add a title
plt.title('Sepal Length Vs. Petal Width')

#Display the chart
plt.show()
```

```python
#Select only numerical columns(ignore the text 'species' column)
numerical_data = df_cleaned.select_dtypes(include='number')

#make the chart a bit larger
plt.figure(figsize=(8,6))

#Create a heatmap with correlation numbers
sns.heatmap(numerical_data.corr(), annot=True, cmap='Blues')

#Add a title
plt.title('Correlation Heatmap')

#Display the chart
plt.show()
```
